```python
def count_radiology_studies(file_path):
    # Read the CSV file contents
    data = pd.read_csv(file_path)

    # Create a dictionary to count the radiology studies frequency
    study_count = {}

    # Iterate over each row in the CSV file and count the occurrences of each radiology study
    for index, row in data.iterrows():
        study_type = row[4]

        if study_type in study_count:
            study_count[study_type] += 1
        else:
            study_count[study_type] = 1

    # Sort the dictionary by values in descending order and get only top 20
    sorted_study_count = {k: v for k, v in sorted(study_count.items(), key=lambda item: item[1], reverse=True)[:20]}

    # Print the result
    print("Radiology Study Count:")
    for study, count in sorted_study_count.items():
        print(f"{study}: {count}")

    # Create a new DataFrame which only includes reports for the top 20 study types
    filtered_data = data[data['Study'].isin(sorted_study_count.keys())]

    # Save the filtered data to a new CSV
    filtered_data.to_csv("/Users/KTang/Desktop/filtered_radiology_studies.csv", index=False)

    return filtered_data

file_path = "/Users/KTang/Desktop/rads_deid_2023_06_08.csv"
count_radiology_studies(file_path)
    
```
